```python
import numpy as np
import pandas as pd
from imageai.Detection import ObjectDetection
import cv2
import sys
import re
import os ,time
import pyttsx3
from PIL import Image
```

```python
model_path = 'models/resnet50_coco_best_v2.0.1.h5'
model_speed = 'fast'
out_path = 'model_output/'
resize_path = 'resizedimages/'
pd_threshould = 30 
detect = Detector(model_path,model_speed,out_path,threshould=pd_threshould)
detect
```

```python
class Detector:
    def __init__(self,modelpath,speed,out_path,threshould=30):
        self.model_path  = modelpath
        self.model_speed = speed
        self.minProbablilityPercent = threshould
        self.output = out_path
        self.engine = pyttsx3.init()
    
    def detect_image(self,image_path):
        self.base_file = os.path.basename(image_path)
        self.detector = ObjectDetection()
        self.define_ModelAttributes()
        self.detections = self.detector.detectCustomObjectsFromImage \
            (input_image=image_path,
             output_image_path= os.path.join(self.output,self.base_file),
             custom_objects=self.custom_objects,
             minimum_percentage_probability=self.minProbablilityPercent,
            extract_detected_objects=False)
        for objects in self.detections:
            obj_name = objects['name']
            print(obj_name)
        
            
    def define_ModelAttributes(self):
        # Define Model Attributes
        self.detector.setModelTypeAsRetinaNet()
        self.detector.setModelPath(self.model_path)
        self.detector.loadModel(detection_speed=self.model_speed)
        self.custom_objects = self.detector.CustomObjects\
            (person=True,
             car=True,
             bench=True,
             sports_ball=True,
             chair=True
             )
        
    def count_detections(self):
        self.person = 0
        self.car = 0
        self.bench = 0
        self.sports_ball =0
        self.chair =0
        self.ObjectList = []
        for objects in self.detections:
            objname = objects['name']
            if objname == 'person':
                self.person += 1
            elif objname == 'bench':
                self.bench += 1
            elif objname == 'car':
                self.car += 1
            elif objname =='sports_ball':
                self.sports_ball +=1
            elif objname =='chair':
                self.chair +=1
        detections_dict = {'image':self.base_file,
                          'Persons':self.person,
                          'car:':self.car,
                          'bench':self.bench,
                          'sports_ball':self.sports_ball,
                          'chair':self.chair}
        return detections_dict

    
def resize_images(input_images,out_path)
    size = (800, 600)
    for i,imgs in enumerate(input_images):
        img_path = os.path.join(input_path,imgs)
        img = Image.open(img_path)
        img = img.resize(size, Image.ANTIALIAS)
        img.save(os.path.join(out_path,'image_'+str(i)+'.jpg'))
    return out_path
    
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
input_images=  os.listdir(input_path)
input_images.sort(key=natural_keys)

for imgpath in input_images:
    img = os.path.join(input_path,imgpath)
    print ("Runnnig image {}".format(img))
    detect.detect_image(img)
    detect_dict = detect.count_detections()
    time.sleep(2)
```
